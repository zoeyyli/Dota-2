```python
# Initialize dataframe and index
df = []

# Initialize keys
keys = ['barracks_status_dire', 'barracks_status_radiant', 'dire_score', 'duration', 'first_blood_time', 'radiant_score', 'radiant_win', 'start_time', 'tower_status_radiant']
pkeys = ['match_id', 'player_slot', 'ability_upgrades_arr', 'account_id', 'assists', 'deaths', 'denies', 'gold_per_min', 'gold_spent', 'hero_damage', 'hero_healing', 'hero_id', 'item_0', 'item_1', 'item_2', 'item_3', 'item_4', 'item_5', 'item_neutral', 'kills', 'last_hits', 'personaname', 'level', 'net_worth', 'tower_damage', 'xp_per_min', 'isRadiant', 'win', 'kills_per_min', 'kda', 'total_gold', 'total_xp']

for i in range(len(match_responses)):
    # Convert match to json
    match = match_responses[i].json()

    # Obtain general match values
    drow = {}
    for key in keys:
        try:
            drow[key] = match[key]
        except:
            drow[key] = None

    # Obtain player data
    players = match['players']

    # Iterate through all player data, creating row in df
    for j in range(10):
        
        # Copy data row
        row = drow.copy()

        # Obtain player values
        prow = {}
        for pkey in pkeys:
            try:
                prow[pkey] = players[j][pkey]
            except:
                prow[pkey] = None

        # Merge data
        row.update(prow)
        # Add row to df
        df.append(row)

```

```python
df = pd.DataFrame(df)

compression_opts = dict(method='zip', archive_name ='addata.csv')
df.to_csv('addata.zip', index = False, compression=compression_opts)
```
